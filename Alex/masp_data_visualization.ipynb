# MASP vs CU Graduation and Retention Rates

## 1st to 2nd Year Retention Rates

```python
masp_ret_rates = [95,88,96,96,94,100,94,92,85,94,94,86,100,93,100,90,88,92,96,94,89,87,95,88,89] # 93'-17'
cu_ret_rates_any = [71,66,77,76,76,80,76,77,77,81,82,76,79,75,76,76,80,79,79,78,80,80,81,82,84] # 93'-17'
cu_ret_rates_as = [65,64,74,74,71,79,75,75,75,80,79,75,78,75,74,73,79,76,75,74,77,74,71,74,76]

masp_ret_years = list(range(1993, 2018)) #ranges don't include right endpoint
cu_ret_years = list(range(1993, 2018))

print("Any college average" , np.mean(cu_ret_rates_any), '\n', "A&S college average", np.mean(cu_ret_rates_as))
```

```python
plt.bar(masp_ret_years[15:], masp_ret_rates[15:], color = "gold", label = "MASP", alpha = .8)
plt.bar(cu_ret_years[15:], cu_ret_rates_as[15:], color = "black",  label = "CU", width = .45, alpha = .9)
plt.legend()
plt.title("1st to 2nd year retention rate - A&S grads")
plt.ylabel("Retention rate")
```

```python
plt.plot(masp_ret_years, masp_ret_rates, label = "MASP", color = "gold")
plt.plot(cu_ret_years, cu_ret_rates_as, label = "CU A&S", color = "black")
plt.ylabel("Retention rate")
plt.title("1st to 2nd year retention rate - A&S grads")
plt.legend()
```

```python
plt.plot(masp_ret_years, masp_ret_rates, label = "MASP", color = "gold")
plt.plot(cu_ret_years, cu_ret_rates_any, label = "CU A&S", color = "black")
plt.ylabel("Retention rate")
plt.title("1st to 2nd year retention rate - CU grads")
plt.legend()
```
