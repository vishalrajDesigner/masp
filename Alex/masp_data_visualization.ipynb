# MASP vs CU Graduation and Retention Rates

## 1st to 2nd Year Retention Rates

```python
masp_ret_rates = [95,88,96,96,94,100,94,92,85,94,94,86,100,93,100,90,88,92,96,94,89,87,95,88,89] # 93'- 17'
cu_ret_rates = [65,64,74,74,71,79,75,75,75,80,79,75,78,75,74,73,79,76,75,74,77,74,71,74,76] # 93' - 17'

ret_years = list(range(1993, 1993+len(masp_ret_rates)))
```

```python
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.bar(ret_years[15:], masp_ret_rates[15:], color = "gold", label = "MASP", alpha = .8)
plt.bar(ret_years[15:], cu_ret_rates[15:], color = "black",  label = "CU", width = .45, alpha = .9)
plt.legend()

plt.title("1st to 2nd year retention rate")
plt.ylabel("Retention rate")

fig.savefig("C:/Users/Alex/masp/Alex/retention_bar.png", bbox_inches = 'tight', format = 'png')
fig.savefig("C:/Users/Alex/masp/Alex/retention_bar.eps", bbox_inches = 'tight', format = 'eps')
```

```python
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.plot(ret_years, masp_ret_rates, label = "MASP", color = "gold", linewidth = 3)
plt.plot(ret_years, cu_ret_rates,  label = "CU A&S", color = "black", linewidth = 3)
plt.ylabel("Retention rate")
plt.title("1st to 2nd year retention rate")
plt.legend()

fig.savefig("C:/Users/Alex/masp/Alex/retention_line.png", bbox_inches = 'tight', format = 'png')
fig.savefig("C:/Users/Alex/masp/Alex/retention_line.eps", bbox_inches = 'tight', format = 'eps')
```

## 6 Year Graduation Rates

```python

masp_grad_rates = [76,59,67,67,70,73,73,80,69,85,71,72,82,75,77,78,79,83,71,83] # 93'-12'
cu_grad_rates = [36,31,39,45,43,49,46,48,44,48,46,49,48,50,51,54,58,53,51,52]

grad_years = list(range(1993, 1993+len(masp_grad_rates)))
```

```python
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xticks(grad_years[1::2])
plt.plot(grad_years, masp_grad_rates, label = "MASP", color = "gold",linewidth = 3)
plt.plot(grad_years, cu_grad_rates, label = "CU A&S", color = "black",linewidth = 3)
plt.ylabel("Graduation rate")
plt.title("6 year graduation rate")
plt.legend()
fig.savefig("C:/Users/Alex/masp/Alex/grad_line.png", bbox_inches = 'tight', format = 'png')
fig.savefig("C:/Users/Alex/masp/Alex/grad_line.eps", bbox_inches = 'tight', format = 'eps')
```
