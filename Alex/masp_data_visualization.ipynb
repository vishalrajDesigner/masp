```python
import matplotlib.pyplot as plt
import numpy as np
```

# MASP vs CU Graduation and Retention Rates

## 1st to 2nd Year Retention Rates

```python
masp_ret_rates = [95,88,96,96,94,100,94,92,85,94,94,86,100,93,100,90,88,92,96,94,89,87,95,88,89] # 93'-17'
cu_ret_rates_any = [71,66,77,76,76,80,76,77,77,81,82,76,79,75,76,76,80,79,79,78,80,80,81,82,84] # 93'-17'
cu_ret_rates_as = [65,64,74,74,71,79,75,75,75,80,79,75,78,75,74,73,79,76,75,74,77,74,71,74,76]

masp_ret_years = list(range(1993, 2018)) #ranges don't include right endpoint
cu_ret_years = list(range(1993, 2018))

print("Any college average" , np.mean(cu_ret_rates_any), '\n', "A&S college average", np.mean(cu_ret_rates_as))
```

```python
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.bar(masp_ret_years[15:], masp_ret_rates[15:], color = "gold", label = "MASP", alpha = .8)
plt.bar(cu_ret_years[15:], cu_ret_rates_as[15:], color = "black",  label = "CU", width = .45, alpha = .9)
plt.legend()

plt.title("1st to 2nd year retention rate - A&S grads")
plt.ylabel("Retention rate")

fig.savefig("C:/Users/Alex/masp/Alex/retention_bar.png", bbox_inches = 'tight', format = 'png')
```

```python
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.plot(masp_ret_years, masp_ret_rates, label = "MASP", color = "gold", linewidth = 3)
plt.plot(cu_ret_years, cu_ret_rates_as,  label = "CU A&S", color = "black", linewidth = 3)
plt.ylabel("Retention rate")
plt.title("1st to 2nd year retention rate - A&S grads")
plt.legend()

fig.savefig("C:/Users/Alex/masp/Alex/retention_line.png", bbox_inches = 'tight', format = 'png')
```

## 6 Year Graduation Rates

```python

masp_grad_rates = [76,59,67,67,70,73,73,80,69,85,71,72,82,75,77,78,79,83,71,83] # 93'-12'
cu_grad_rates_any = [44,38,48,50,52,57,58,54,52,53,55,56,54,55,57,60,63,60,56,60]
cu_grad_rates_as = [36,31,39,45,43,49,46,48,44,48,46,49,48,50,51,54,58,53,51,52]

masp_grad_years = list(range(1993, 2013))
cu_grad_years = list(range(1993, 2013))

print("Any college average" , np.mean(cu_grad_rates_any), '\n', "A&S college average", np.mean(cu_grad_rates_as))
```

```python
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xticks(masp_grad_years[1::2])
plt.plot(masp_grad_years, masp_grad_rates, label = "MASP", color = "gold",linewidth = 3)
plt.plot(cu_grad_years, cu_grad_rates_as, label = "CU A&S", color = "black",linewidth = 3)
plt.ylabel("Graduation rate")
plt.title("6 year graduation rate - A&S grads")
plt.legend()
fig.savefig("C:/Users/Alex/masp/Alex/grad_line.png", bbox_inches = 'tight', format = 'png')
```
