# MASP vs CU Graduation and Retention Rates

## 1st to 2nd Year Retention Rates

```python
import matplotlib.pyplot as plt
import os

filepath = os.getcwd() + '\\'
```

```python
ret_years = list(range(1993, 1993+len(masp_ret_rates)))
```

```python
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.plot(ret_years, masp_ret_rates, color = "black", linewidth = 3)
plt.plot(ret_years, cu_ret_rates, color = "gold", linewidth = 3)
plt.ylabel("Retention rate", fontsize = 14)

fig.text(.15,.80, "MASP", color = "black", fontsize = 16)
fig.text(.15,.48, "CU Arts & Sciences", color = "gold", fontsize = 15)

print(filepath + "retention_line.png")
fig.savefig(filepath + "retention_line.png", bbox_inches = 'tight', format = 'png')
fig.savefig(filepath + "retention_line.eps", bbox_inches = 'tight', format = 'eps')
```

```python
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

offset = len(ret_years) - years_to_show

plt.bar(ret_years[offset:], masp_ret_rates[offset:], color = "gold", alpha = .8)
plt.bar(ret_years[offset:], cu_ret_rates[offset:], color = "black", width = .45, alpha = .9)

plt.ylabel("Retention rate", fontsize = 14)

fig.savefig(filepath + "retention_bar.png", bbox_inches = 'tight', format = 'png')
fig.savefig(filepath + "retention_bar.eps", bbox_inches = 'tight', format = 'eps')
```

## 6 Year Graduation Rates

```python
grad_years = list(range(1993, 1993+len(masp_grad_rates)))
```

```python
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xticks(grad_years[1::2])
plt.plot(grad_years, masp_grad_rates, label = "MASP", color = "black",linewidth = 3)
plt.plot(grad_years, cu_grad_rates, label = "CU A&S", color = "gold",linewidth = 3)
plt.ylabel("Graduation rate", fontsize = 14)
fig.text(.2,.70, "MASP", color = "black", fontsize = 16)
fig.text(.2,.42, "CU Arts & Sciences", color = "gold", fontsize = 15)

fig.savefig(filepath + "grad_line.png", bbox_inches = 'tight', format = 'png')
fig.savefig(filepath + "grad_line.eps", bbox_inches = 'tight', format = 'eps')
```

```python
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

offset = len(grad_years) - years_to_show

plt.bar(grad_years[offset:], masp_grad_rates[offset:], color = "gold", alpha = .8)
plt.bar(grad_years[offset:], cu_grad_rates[offset:], color = "black", width = .45, alpha = .9)

plt.ylabel("Graduation rate", fontsize = 14)

fig.savefig(filepath + "grad_bar.png", bbox_inches = 'tight', format = 'png')
fig.savefig(filepath + "grad_bar.eps", bbox_inches = 'tight', format = 'eps')
```
